### Plotting Trigonometry Graph in Python

first we imported the numpy and matplotlib in python 

```python
import numpy as np 
import matplotlib.pyplot as plt
```

then we used the linspace to make add the numbers inside the array

```python
x = np.linspace(0, 2*np.pi, num = 100)
```

```python
y = np.sin(x)
```

then we plotted the graph using matplotlib in python

```python
plt.plot(y)
```
